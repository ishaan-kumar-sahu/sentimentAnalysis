### Lemmatization

```python
from nltk.stem import WordNetLemmatizer

# initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# define a function to perform lemmatization using NLTK
def lemmatize_text(text):
    if isinstance(text, str):
        tokens = nltk.word_tokenize(text)
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
        lemmatized_text = ' '.join(lemmatized_tokens)
        return lemmatized_text
    else:
        return ''

# apply the lemmatization function to the dataframe
df['lemmatized_text'] = df['text'].apply(lemmatize_text)

# print the resulting dataframe
print(df)
```

### Removing Stopwords

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# initialize the WordNetLemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
# define a function to tokenize, lemmatize, and remove stop words from text
def process_text(text):
    if isinstance(text, str):
        # tokenize the text
        tokens = word_tokenize(text.lower())

        # lemmatize the tokens
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

        # remove stop words from the lemmatized tokens
        filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words]

        # join the filtered tokens into a string
        filtered_text = ' '.join(filtered_tokens)

        return filtered_text
    else:
        return ''

# apply the text processing function to the text column in the dataframe
df['processed_text'] = df['text'].apply(process_text)

# print the resulting dataframe
print(df)
```

### Pos Tagging

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


# initialize the WordNetLemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# define a function to tokenize, lemmatize, remove stop words, and perform POS tagging on text
def process_text(text):
    if isinstance(text, str):
        # tokenize the text
        tokens = word_tokenize(text.lower())
    # lemmatize the tokens
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

        # remove stop words from the lemmatized tokens
        filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words]

        # perform POS tagging on the filtered tokens
        pos_tags = nltk.pos_tag(filtered_tokens)

        return pos_tags
    else:
        return []

# apply the text processing function to the text column in the dataframe
df['processed_text'] = df['text'].apply(process_text)

# print the resulting dataframe
print(df)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Instantiate Vader sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Create a function to convert the sentiment score into positive, negative, and neutral categories
def categorize_sentiment(score):
    if score > 0.05:
        return 'positive'
    elif score < -0.05:
        return 'negative'
    else:
        return 'neutral'
    
 # Create a function to apply Vader sentiment analysis and convert the sentiment score into categories for each row of dataframe
def analyze_sentiment(row):
    if isinstance(row['text'], str):
        sentiment_scores = sid.polarity_scores(row['text'])
        sentiment_score = sentiment_scores['compound']
        sentiment_category = categorize_sentiment(sentiment_score)
        return sentiment_category
    else:
        return 'neutral'   
    
# Apply the sentiment analysis function to the 'text' column of the dataframe and create a new column for sentiment scores
df['sentiment_score'] = df.apply(analyze_sentiment, axis=1)

# Print the dataframe to see the sentiment scores
print(df)
```

```python
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer


# Instantiate Vader sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Create a function to apply Vader sentiment analysis to each row of dataframe
def analyze_sentiment(row):
    if isinstance(row['text'], str):
        sentiment_scores = sid.polarity_scores(row['text'])
        return sentiment_scores['compound']
    else:
        return 0.0

# Apply the sentiment analysis function to the 'text' column of the dataframe and create a new column for sentiment scores
df['sentiment_score'] = df.apply(analyze_sentiment, axis=1)

# Categorize sentiment scores as positive, negative, or neutral based on a threshold value
threshold = 0.05
df['sentiment_category'] = pd.cut(df['sentiment_score'], bins=[-1, -threshold, threshold, 1], labels=['negative', 'neutral', 'positive'])

# Group the dataframe by sentiment category and count the number of occurrences of each category
sentiment_counts = df.groupby('sentiment_category').size()

# Create a bar plot of the sentiment categories and their corresponding counts
sentiment_counts.plot(kind='bar')

# Set the x-axis label and title for the plot
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Occurrences')
plt.title('Sentiment Analysis Results')

# Display the plot
plt.show()
```
